```python
# %pip install easyocr

import pandas as pd
import easyocr as eo
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
from tqdm import tqdm
```

```python
def extract_text_with_binary_mask(image_path, languages=['en'], gpu=False):
    reader = eo.Reader(languages, gpu=gpu)
    results = reader.readtext(image_path)
    
    image = cv2.imread(image_path)
    height, width, _ = image.shape
    
    binary_mask = np.zeros((height, width), dtype=np.uint8)
    
    for result in results:
        bbox, text, score = result
        l_bbox, l_bbox1 = bbox[0]
        r_bbox, r_bbox1 = bbox[2]
        
        cv2.rectangle(binary_mask, 
                      (int(l_bbox), int(l_bbox1)), 
                      (int(r_bbox), int(r_bbox1)), 
                      255, -1)
    
    return image, binary_mask
```

```python
image, binary_mask = extract_text_with_binary_mask('./dataset/anti-Semitism meme offensive/Image_4.jpg')

text_removed_image = cv2.inpaint(image, binary_mask, inpaintRadius=0.03, flags=cv2.INPAINT_TELEA)

original_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
text_removed_rgb = cv2.cvtColor(text_removed_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(original_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_mask, cmap='gray')
plt.title('Binary Mask')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(text_removed_rgb)
plt.title('Text Removed')
plt.axis('off')

plt.show()
```

```python
def resize_and_center_crop(image, target_min=300, target_max=512, target_size=512):
    height, width, _ = image.shape

    if width < target_min or height < target_min:
        scale_factor = target_min / min(width, height)
        new_width = int(width * scale_factor)
        new_height = int(height * scale_factor)
        image = cv2.resize(image, (new_width, new_height))

    height, width, _ = image.shape
    if width > target_max or height > target_max:
        scale_factor = target_max / max(width, height)
        new_width = int(width * scale_factor)
        new_height = int(height * scale_factor)
        image = cv2.resize(image, (new_width, new_height))

    height, width, _ = image.shape
    center_x, center_y = width // 2, height // 2
    crop_x1 = max(center_x - target_size // 2, 0)
    crop_x2 = min(center_x + target_size // 2, width)
    crop_y1 = max(center_y - target_size // 2, 0)
    crop_y2 = min(center_y + target_size // 2, height)
    cropped_image = image[crop_y1:crop_y2, crop_x1:crop_x2]

    if cropped_image.shape[0] != target_size or cropped_image.shape[1] != target_size:
        cropped_image = cv2.resize(cropped_image, (target_size, target_size))

    return cropped_image
```

```python
# Example usage
processed_image = resize_and_center_crop(text_removed_rgb, target_min=300, target_max=512, target_size=512)

# Display the processed image
plt.imshow(processed_image)
plt.title("Processed Image")
plt.axis('off')
plt.show()
```

```python
def extract_text_with_binary_mask(image_path, languages=['en'], gpu=False):
    reader = eo.Reader(languages, gpu=gpu)
    results = reader.readtext(image_path)
    
    image = cv2.imread(image_path)
    height, width, _ = image.shape
    
    binary_mask = np.zeros((height, width), dtype=np.uint8)
    
    for result in results:
        bbox, text, score = result
        l_bbox, l_bbox1 = bbox[0]
        r_bbox, r_bbox1 = bbox[2]
        
        cv2.rectangle(binary_mask, 
                      (int(l_bbox), int(l_bbox1)), 
                      (int(r_bbox), int(r_bbox1)), 
                      255, -1)
    
    return image, binary_mask

# Function to resize and center crop
def resize_and_center_crop(image, target_min=300, target_max=512, target_size=512):
    height, width, _ = image.shape

    # Resize if the smallest dimension is smaller than target_min
    if width < target_min or height < target_min:
        scale_factor = target_min / min(width, height)
        new_width = int(width * scale_factor)
        new_height = int(height * scale_factor)
        image = cv2.resize(image, (new_width, new_height))

    # Resize if the largest dimension exceeds target_max
    height, width, _ = image.shape
    if width > target_max or height > target_max:
        scale_factor = target_max / max(width, height)
        new_width = int(width * scale_factor)
        new_height = int(height * scale_factor)
        image = cv2.resize(image, (new_width, new_height))

    # Center crop to target_size
    height, width, _ = image.shape
    center_x, center_y = width // 2, height // 2
    crop_x1 = max(center_x - target_size // 2, 0)
    crop_x2 = min(center_x + target_size // 2, width)
    crop_y1 = max(center_y - target_size // 2, 0)
    crop_y2 = min(center_y + target_size // 2, height)
    cropped_image = image[crop_y1:crop_y2, crop_x1:crop_x2]

    if cropped_image.shape[0] != target_size or cropped_image.shape[1] != target_size:
        cropped_image = cv2.resize(cropped_image, (target_size, target_size))

    return cropped_image

# Process images from train_df
def process_and_replace_images(train_df, output_folder, languages=['en'], gpu=False):
    for idx, row in tqdm(train_df.iterrows(), total=len(train_df)):
        image_path = row['new_image_path']
        output_path = os.path.join(output_folder, os.path.basename(image_path))

        try:
            # Extract image and binary mask
            image, binary_mask = extract_text_with_binary_mask(image_path, languages, gpu)

            # Remove text using inpainting
            text_removed_image = cv2.inpaint(image, binary_mask, inpaintRadius=3, flags=cv2.INPAINT_TELEA)

            # Resize and center crop
            processed_image = resize_and_center_crop(text_removed_image, target_min=300, target_max=512, target_size=512)

            # Save the processed image
            cv2.imwrite(output_path, processed_image)

        except Exception as e:
            print(f"Error processing {image_path}: {e}")


train_df = pd.read_csv('final_dataset.csv')  # Load your dataframe
output_folder = './final_dataset/'  # Path to your output folder

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Process all images
process_and_replace_images(train_df, output_folder, languages=['en'], gpu=True)

print("Processing complete!")
```
